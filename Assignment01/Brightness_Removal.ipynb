```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('image.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image (RGB)")
plt.axis('off')
```

# **HSV - Removing V Channel**

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

v_removed = np.zeros_like(v)
hsv_no_brightness = cv2.merge([h, s, v_removed])
result_hsv = cv2.cvtColor(hsv_no_brightness, cv2.COLOR_HSV2RGB)

plt.imshow(result_hsv)
plt.title("HSV without Brightness (V removed)")
plt.axis('off')
```

# **YUV - Removing Y Channel**

```python
yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
y, u, v = cv2.split(yuv)

y_removed = np.zeros_like(y)
yuv_no_brightness = cv2.merge([y_removed, u, v])
result_yuv = cv2.cvtColor(yuv_no_brightness, cv2.COLOR_YUV2RGB)

plt.imshow(result_yuv)
plt.title("YUV without Brightness (Y removed)")
plt.axis('off')
```

# **Lab - Removing L Channel**

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)

l_removed = np.zeros_like(l)
lab_no_brightness = cv2.merge([l_removed, a, b])
result_lab = cv2.cvtColor(lab_no_brightness, cv2.COLOR_LAB2RGB)

plt.imshow(result_lab)
plt.title("Lab without Brightness (L removed)")
plt.axis('off')
```

# **CMYK - Removing K Channel**

```python
rgb = img_rgb.astype(float) / 255.0
K = 1 - np.max(rgb, axis=2)
C = (1 - rgb[:,:,0] - K) / (1 - K + 1e-8)
M = (1 - rgb[:,:,1] - K) / (1 - K + 1e-8)
Y = (1 - rgb[:,:,2] - K) / (1 - K + 1e-8)

K_removed = np.zeros_like(K)

rgb_no_brightness = np.dstack([
    (1 - C) * (1 - K_removed),
    (1 - M) * (1 - K_removed),
    (1 - Y) * (1 - K_removed)
])

plt.imshow(rgb_no_brightness)
plt.title("CMYK without Brightness (K removed)")
plt.axis('off')
```
