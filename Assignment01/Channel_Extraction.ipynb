```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('image.jpg')   # BGR by default
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image (RGB)")
plt.axis('off')
```

# **RGB Color Space**
## *Channels: Red, Green, Blue*

```python
R, G, B = cv2.split(img_rgb)

plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(R, cmap='gray')
plt.title("Red Channel")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(G, cmap='gray')
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(B, cmap='gray')
plt.title("Blue Channel")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(img_rgb)
plt.title("Combined RGB")
plt.axis('off')
```

# **BGR Color Space**
# *Channels: Blue, Green, Red*

```python
B, G, R = cv2.split(img)

plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(B, cmap='gray')
plt.title("Blue Channel")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(G, cmap='gray')
plt.title("Green Channel")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(R, cmap='gray')
plt.title("Red Channel")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Combined BGR")
plt.axis('off')
```

# **HSV Color Space**
# *Channels: Hue, Saturation, Value*

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
H, S, V = cv2.split(hsv)

plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(H, cmap='gray')
plt.title("Hue Channel")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(S, cmap='gray')
plt.title("Saturation Channel")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(V, cmap='gray')
plt.title("Value Channel")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB))
plt.title("Combined HSV")
plt.axis('off')
```

# **CMYK Color Space**
# *Channels: Cyan, Magenta, Yellow, Black*

```python
import numpy as np

rgb = img_rgb.astype(float) / 255.0
K = 1 - np.max(rgb, axis=2)
C = (1 - rgb[:,:,0] - K) / (1 - K + 1e-8)
M = (1 - rgb[:,:,1] - K) / (1 - K + 1e-8)
Y = (1 - rgb[:,:,2] - K) / (1 - K + 1e-8)

plt.figure(figsize=(15,4))

plt.subplot(1,5,1)
plt.imshow(C, cmap='gray')
plt.title("Cyan")

plt.subplot(1,5,2)
plt.imshow(M, cmap='gray')
plt.title("Magenta")

plt.subplot(1,5,3)
plt.imshow(Y, cmap='gray')
plt.title("Yellow")

plt.subplot(1,5,4)
plt.imshow(K, cmap='gray')
plt.title("Black")

plt.subplot(1,5,5)
plt.imshow(img_rgb)
plt.title("Combined CMYK")
plt.axis('off')
```

# **YUV Color Space**
# *Channels: Y(Luminance), U, V*

```python
yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
Y, U, V = cv2.split(yuv)

plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(Y, cmap='gray')
plt.title("Y Channel")

plt.subplot(1,4,2)
plt.imshow(U, cmap='gray')
plt.title("U Channel")

plt.subplot(1,4,3)
plt.imshow(V, cmap='gray')
plt.title("V Channel")

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(yuv, cv2.COLOR_YUV2RGB))
plt.title("Combined YUV")
plt.axis('off')
```

# **Lab Color Space**
# *Channels: L(Light), a, b*

```python
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
L, A, B = cv2.split(lab)

plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(L, cmap='gray')
plt.title("L Channel")

plt.subplot(1,4,2)
plt.imshow(A, cmap='gray')
plt.title("a Channel")

plt.subplot(1,4,3)
plt.imshow(B, cmap='gray')
plt.title("b Channel")

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(lab, cv2.COLOR_LAB2RGB))
plt.title("Combined Lab")
plt.axis('off')
```
